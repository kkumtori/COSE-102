```python
region=["중구","진관동","서울시","서대문구","강남구"]

count_gang=0
count_gu=0

for each_region in region:
    if(each_region[0]=="강"): # 강으로 시작하는 행정구역
        count_gang=count_gang+1  
    if(each_region[-1]=="구"): # 구로 끝나는 행정구역
        count_gu=count_gu+1

print("강으로 시작 : ",count_gang,"회")
print("구로 끝 : ", count_gu,"회")
```

```python
print("세일 전 품복별 금액을 아래에 입력하세요.") 
price=input().split()

new_price=[]
for item in price:
    new_price.append(int(item)//2)
    # 지칭 대명사 item 은 포인터 대상이 되는 값을 변경하는 것이 불가능능

print(new_price)
```

```python
# Solution 2 
print("세일 전 품목별 금액을 아래에 입력하세요.")
price=input().split()

i=0

while(i<len(price)):
    price[i]=int(price[i]) # 재정의
    price[i]=price[i]//2
    i=i+1

print(price)
```

```python
# Solution 3  
print("세일 전 품복별 금액을 아래에 입력하세요.") 
price=input().split()

for index in range(0,len(price)):
    price[index]=int(price[index])//2

print(price)
```

## 데이터 정렬 방법론  

### Bubble Sort Algorithm   
len(list)-1 회전 필요  

+ 1회전  
7-4-5-1-3  
4-7-5-1-3  
4-5-7-1-3  
4-5-1-7-3  
4-5-1-3-7  
+ 2회전  
4-1-5-3-7  
4-1-3-5-7  
.  
.  
.  


```python
num=[8,3,4,2,6,1,9,5,7]
#   [0,1,2,3,4,5,6,7,8]

i=1 #회전수
length= len(num)

while(i<length):
    j=0
    while(j<length-i):
        #i=1 일때 [7],[8]까지 비교 
        #i=2 일때 [6],[7]까지 비교교
        if(num[j]>num[j+1]):
            num[j],num[j+1]=num[j+1],num[j] #swap
        j=j+1
    i=i+1

print(num)
```

### Insertion Sort Algorithm  
: 삽입 정렬  
[8,5,6,2,4]일때  
len(list)-1 회전 필요  



```python
num=[8,3,4,2,6,1,9,5,7]

i=1 # 회전수, key값의 index 
length=len(num)

while(i<length): # i: 1~마지막 index
    j=i-1        # j: 0~마지막-1 index
    key_value=num[i]
    while(j>=0):
        if(key_value<num[j]):
            num[j+1]=num[j]
        else:
            break
        j=j-1
    num[j+1]=key_value  # j=0 or 끊긴 곳곳   
    i=i+1
        
print(num)
```

### Selection Sort Algorithm  
: 선택 정렬  
min or max를 찾아 끝으로 이동  


```python
num=[8,3,4,2,6,1,9,5,7]

i=0
length=len(num)

while(i<length-1):
    targets=num[i:length] # [i부터 마지막 인덱스]
    Min=min(targets)
    Index=num.index(Min)
    num[i],num[Index]=num[Index],num[i]
    i=i+1

print(num)
```

## Class  

```python
class Candidate: #Class 정의

    # class 속성
    region='서울 성북 병'

    # 생성자 : class가 생성될때 instance 속성을 포함하는 메소드
    def __init__(self,name,gender,age,number,rating):
        self.name=name
        self.gender=gender
        self.age=age
        self.number=number
        self.rating=rating

    # 소멸자 : class가 소멸될 때 호출되는 메소드
    def __del__(self):
        print(self.name+"후보자 사퇴하였습니다.")

    # 정보 공개 메소드
    def Print_Info(self):
        print("후보자 정보 공개")
        print("이름 :",self.name)
        if(self.gender=="M"):
            print("성별 : 남성")
        else:
            print("성별 : 여성")
        print("나이 :",self.age,"세")

    # 지지율 메소드
    def Print_Rating(self):
        print("기호",self.number,"번",self.name,"후보자 지지율 공개합니다.")
        print(self.rating,"%")

    # 캠페인 메소드
    def Campaign(self,num):
        print("유세를 총",num,"회 실시하였습니다.")
        print("기호",self.number,"번",self.name,"파이팅!")
        print("지지율이",num,"%p 상승했습니다.")
        self.rating=self.rating+num

    def Bad_New(self,num):
        print("아이고! 악재가",num,"번 발생했습니다.")
        print("지지율이",num,"%p하락했습니다.")
        self.rating=self.rating-num

    # 객체간 상호작용
    def Unite(self,other):
        print(self.name,"후보와",other.name,"후보가 단일화하였습니다.")
        sum_rating=self.rating+other.rating
        self.rating=sum_rating
            # 메소드 내 지역변수 설정 가능
        print(self.name,"후보의 지지율은",sum_rating,"%p입니다")
```

```python
text='액션이 어떻게 되는지 아세요? 무한! 무야~호~'
filter=['~','!','@','*','%','^','&','?']

for i in filter:
    text=text.replace(i,'')

print(text)
```

## Tokenization  
+ split() : 어절 단위 토큰화  
+ Komoran : 형태소 단위  

```python
# split()
text_token=text.split()
print(text_token)
```

```python
# morphs
from konlpy.tag import Komoran
komoran=Komoran()

result=komoran.morphs(text)
print(result)
```

```python
# pos
pos_result=komoran.pos(text)
print(pos_result)
```

```python
# noun
noun_result=komoran.nouns(text)
print(noun_result)
```

## TF-IDF와 문서 유사도  
: Term Frequency-Inverse Document Frequency  
+ $\frac{특정~문서~낱말빈도}{낱말등장~문서~수}$  
+ 낱말의 등장 횟수, 낱말 등장한 문서의 희소성 고려  
    + 신문기사에서 모든 문서에 '-기자'는 항상 등장함  
    + '-기자'는 TF-IDF 값이 낮음
+ 자주 등장하는 낱말일수록 중요도 높다는 전제  


```python
# 전처리 완료된 기사 제목 리스트  
articles=['대선 후보 전략 분석',
          '경제 위기 신호 감지',
          '주식 시장 혼란 전략',
          '대선 여론 조사 역전 조사 결과 분석']

from sklearn.feature_extraction.text import CountVectorizer

vector=CountVectorizer()
print(vector.fit_transform(articles).toarray())
print(vector.vocabulary_)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf_idf=TfidfVectorizer().fit(articles)
tf_idf_matrix=tf_idf.transform(articles).toarray()
print(tf_idf_matrix)
print(tf_idf.vocabulary_)
```
